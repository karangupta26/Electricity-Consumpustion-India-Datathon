```python
df['Dates']=pd.to_datetime(df['Dates'])
df['Dates'].describe()
```

```python
df=df[(df['Dates'] >="2019-12-01") & (df['Dates'] <= "2020-12-05")]
df
```

```python
df.loc[(df['Dates'] <="2020-03-24") & (df['Dates'] >= "2019-01-12"),'Situation']='Pre-Covid'
```

```python
df.loc[(df['Dates'] >="2020-03-25") & (df['Dates'] <= "2020-12-05"),'Situation']='Covid-Lockdown'
```

```python
def getSeason(date,index):
    month = int(str(date).replace(" 00:00:00",'').split("-")[1])
    if (month > 11 or month <= 3):
        df['Season'][index]="WINTER"
    elif (month == 4 or month == 5):
        df['Season'][index]="SPRING"
    elif (month >=6 and month <= 9):
        df['Season'][index]="SUMMER"
    else:
        df['Season'][index]="FALL"
```

```python
df['Season']=None
for i,row in df.iterrows():
    getSeason(row['Dates'],i)
```
